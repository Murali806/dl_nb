## Why is it Negative?

```
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”

```

````python
def softmax(x):
    """Numerically stable softmax"""
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

def softmax_derivative_i_not_j(S_i, S_j):
    """Derivative when iâ‰ j: -S_i Ã— S_j"""
    return -S_i * S_j

# Example calculation
logits = np.array([2.0, 1.0, 0.1])
probs = softmax(logits)
classes = ['Cat', 'Dog', 'Mouse']

print('SOFTMAX DERIVATIVE (CASE iâ‰ j)')
print('='*70)
print(f'Logits: {logits}')
print(f'Probabilities: {probs}\n')

print('Cross-term derivatives (âˆ‚S_i/âˆ‚L_j where iâ‰ j):')
print('-'*70)
for i in range(len(probs)):
    for j in range(len(probs)):
        if i != j:
            deriv = softmax_derivative_i_not_j(probs[i], probs[j])
            print(f'âˆ‚S_{i}/âˆ‚L_{j} ({classes[i]}/{classes[j]}): '
                  f'-{probs[i]:.4f} Ã— {probs[j]:.4f} = {deriv:.4f}')

print('='*70)
print('OBSERVATION:')
print('  All cross-term derivatives are NEGATIVE!')
print('  Increasing one class probability decreases others.')
````

## Building the Jacobian Step-by-Step

```
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚         JACOBIAN CONSTRUCTION                               â”‚
â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚                                                             â”‚
â”‚  Given: S = [S_0, S_1, S_2] = [0.65, 0.25, 0.10]          â”‚
â”‚                                                             â”‚
â”‚  Step 1: Compute diagonal terms (i=j)                      â”‚
â”‚  â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”                 â”‚
â”‚  â”‚  J[0,0] = S_0(1-S_0)                 â”‚                 â”‚
â”‚  â”‚         = 0.65 Ã— 0.35 = 0.2275       â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  J[1,1] = S_1(1-S_1)                 â”‚                 â”‚
â”‚  â”‚         = 0.25 Ã— 0.75 = 0.1875       â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  J[2,2] = S_2(1-S_2)                 â”‚                 â”‚
â”‚  â”‚         = 0.10 Ã— 0.90 = 0.0900       â”‚                 â”‚
â”‚  â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜                 â”‚
â”‚                                                             â”‚
â”‚  Step 2: Compute off-diagonal terms (iâ‰ j)                  â”‚
â”‚  â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”                 â”‚
â”‚  â”‚  J[0,1] = -S_0 Ã— S_1                 â”‚                 â”‚
â”‚  â”‚         = -0.65 Ã— 0.25 = -0.1625     â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  J[0,2] = -S_0 Ã— S_2                 â”‚                 â”‚
â”‚  â”‚         = -0.65 Ã— 0.10 = -0.0650     â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  J[1,0] = -S_1 Ã— S_0                 â”‚                 â”‚
â”‚  â”‚         = -0.25 Ã— 0.65 = -0.1625     â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  J[1,2] = -S_1 Ã— S_2                 â”‚                 â”‚
â”‚  â”‚         = -0.25 Ã— 0.10 = -0.0250     â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  J[2,0] = -S_2 Ã— S_0                 â”‚                 â”‚
â”‚  â”‚         = -0.10 Ã— 0.65 = -0.0650     â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  J[2,1] = -S_2 Ã— S_1                 â”‚                 â”‚
â”‚  â”‚         = -0.10 Ã— 0.25 = -0.0250     â”‚                 â”‚
â”‚  â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜                 â”‚
â”‚                                                             â”‚
â”‚  Step 3: Assemble complete matrix                          â”‚
â”‚         â”Œ                                â”                 â”‚
â”‚         â”‚  0.2275  -0.1625  -0.0650     â”‚                 â”‚
â”‚    J =  â”‚ -0.1625   0.1875  -0.0250     â”‚                 â”‚
â”‚         â”‚ -0.0650  -0.0250   0.0900     â”‚                 â”‚
â”‚         â””                                â”˜                 â”‚
â”‚                                                             â”‚
â”‚  Properties:                                               â”‚
â”‚  â€¢ Symmetric: J[i,j] = J[j,i] (when iâ‰ j)                  â”‚
â”‚  â€¢ Diagonal positive, off-diagonal negative                â”‚
â”‚  â€¢ Each row sums to 0 (probability constraint!)            â”‚
â”‚                                                             â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
```

````python
def softmax_jacobian(probs):
    """
    Compute complete Jacobian matrix for softmax.
    
    Args:
        probs: Softmax probabilities (n,)
    
    Returns:
        Jacobian matrix (n, n)
    """
    n = len(probs)
    jacobian = np.zeros((n, n))
    
    for i in range(n):
        for j in range(n):
            if i == j:
                # Diagonal: S_i(1 - S_i)
                jacobian[i, j] = probs[i] * (1 - probs[i])
            else:
                # Off-diagonal: -S_i Ã— S_j
                jacobian[i, j] = -probs[i] * probs[j]
    
    return jacobian

# Compute Jacobian
J = softmax_jacobian(probs)

print('COMPLETE JACOBIAN MATRIX')
print('='*70)
print(f'Probabilities: {probs}\n')
print('Jacobian J[i,j] = âˆ‚S_i/âˆ‚L_j:')
print(J)
print('\n' + '='*70)

# Verify properties
print('\nPROPERTIES VERIFICATION:')
print('-'*70)
print(f'1. Diagonal elements (positive):')
for i in range(len(probs)):
    print(f'   J[{i},{i}] = {J[i,i]:.4f}')

print(f'\n2. Symmetry (J[i,j] = J[j,i] for iâ‰ j):')
print(f'   J[0,1] = {J[0,1]:.4f}, J[1,0] = {J[1,0]:.4f} âœ“')
print(f'   J[0,2] = {J[0,2]:.4f}, J[2,0] = {J[2,0]:.4f} âœ“')
print(f'   J[1,2] = {J[1,2]:.4f}, J[2,1] = {J[2,1]:.4f} âœ“')

print(f'\n3. Row sums (should be 0):')
for i in range(len(probs)):
    row_sum = np.sum(J[i, :])
    print(f'   Row {i}: {row_sum:.10f}')

print('='*70)
````

# Part 4: Using the Jacobian in Backpropagation

## The Chain Rule

```
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚         JACOBIAN IN BACKPROPAGATION                         â”‚
â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚                                                             â”‚
â”‚  During backpropagation, we need:                          â”‚
â”‚                                                             â”‚
â”‚  âˆ‚Loss/âˆ‚L = âˆ‚Loss/âˆ‚S Ã— âˆ‚S/âˆ‚L                              â”‚
â”‚              â†‘           â†‘                                  â”‚
â”‚              â”‚           â””â”€ Jacobian matrix                â”‚
â”‚              â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€ Gradient from loss             â”‚
â”‚                                                             â”‚
â”‚  In matrix form:                                           â”‚
â”‚  â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”                 â”‚
â”‚  â”‚  dL/dL = J^T @ dL/dS                 â”‚                 â”‚
â”‚  â”‚                                       â”‚                 â”‚
â”‚  â”‚  Where:                               â”‚                 â”‚
â”‚  â”‚  â€¢ dL/dS: gradient w.r.t. softmax    â”‚                 â”‚
â”‚  â”‚           (n Ã— 1 vector)              â”‚                 â”‚
â”‚  â”‚  â€¢ J^T: transposed Jacobian           â”‚                 â”‚
â”‚  â”‚         (n Ã— n matrix)                â”‚                 â”‚
â”‚  â”‚  â€¢ dL/dL: gradient w.r.t. logits     â”‚                 â”‚
â”‚  â”‚           (n Ã— 1 vector)              â”‚                 â”‚
â”‚  â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜                 â”‚
â”‚                                                             â”‚
â”‚  Example (3 classes):                                      â”‚
â”‚                                                             â”‚
â”‚  dL/dS = [0.1, -0.5, 0.2]  (from loss function)           â”‚
â”‚                                                             â”‚
â”‚         â”Œ                    â”   â”Œ      â”                  â”‚
â”‚         â”‚  0.23  -0.16  -0.07â”‚   â”‚  0.1 â”‚                  â”‚
â”‚  J^T =  â”‚ -0.16   0.19  -0.03â”‚ @ â”‚ -0.5 â”‚                  â”‚
â”‚         â”‚ -0.07  -0.03   0.09â”‚   â”‚  0.2 â”‚                  â”‚
â”‚         â””                    â”˜   â””      â”˜                  â”‚
â”‚                                                             â”‚
â”‚  dL/dL = [result of matrix multiplication]                â”‚
â”‚                                                             â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
```

````python
# Example: Backpropagation through softmax
def backprop_through_softmax(dL_dS, probs):
    """
    Backpropagate gradient through softmax layer.
    
    Args:
        dL_dS: Gradient w.r.t. softmax outputs (n,)
        probs: Softmax probabilities (n,)
    
    Returns:
        dL_dL: Gradient w.r.t. logits (n,)
    """
    # Compute Jacobian
    J = softmax_jacobian(probs)
    
    # Apply chain rule: dL/dL = J^T @ dL/dS
    dL_dL = J.T @ dL_dS
    
    return dL_dL

# Example gradient from loss
dL_dS = np.array([0.1, -0.5, 0.2])

# Backpropagate
dL_dL = backprop_through_softmax(dL_dS, probs)

print('BACKPROPAGATION EXAMPLE')
print('='*70)
print(f'Softmax probabilities: {probs}')
print(f'\nGradient from loss (dL/dS): {dL_dS}')
print(f'\nJacobian matrix:')
print(J)
print(f'\nGradient w.r.t. logits (dL/dL): {dL_dL}')
print('='*70)

# Verify with element-wise computation
print('\nVERIFICATION (element-wise):')  
for i in range(len(probs)):
    manual = sum(J[j, i] * dL_dS[j] for j in range(len(probs)))
    print(f'  dL/dL_{i} = {manual:.4f} (matches {dL_dL[i]:.4f})')
````
