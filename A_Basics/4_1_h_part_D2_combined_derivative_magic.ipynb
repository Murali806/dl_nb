# Part D2: The Magic Combined Derivative

**StatQuest Style**: "The most beautiful result in deep learning... clearly!"

### The Magic: âˆ‚L/âˆ‚x = y_pred - y_true

````python
import numpy as np

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / np.sum(exp_x)

def combined_gradient(y_true, y_pred):
    """Combined softmax + cross entropy gradient"""
    return y_pred - y_true

# Example
logits = np.array([2.0, 1.0, 0.1])
y_true = np.array([1.0, 0.0, 0.0])  # Cat
y_pred = softmax(logits)

gradient = combined_gradient(y_true, y_pred)

print('THE MAGIC COMBINED DERIVATIVE')
print('='*60)
print(f'True: {y_true}')
print(f'Pred: {y_pred}')
print(f'\nGradient = Pred - True')
print(f'         = {gradient}')
print('='*60)
print('\nThis simple formula is ready for backpropagation!')
print('All the complex math cancels out beautifully!')
````
