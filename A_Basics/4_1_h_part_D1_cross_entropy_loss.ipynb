# Part D1: Cross Entropy Loss

**StatQuest Style**: "Now let's measure how wrong we are... clearly!"

### Formula: L = -Î£ y_true Ã— log(y_pred)

````python
import numpy as np

def cross_entropy_loss(y_true, y_pred):
    """Cross entropy loss"""
    epsilon = 1e-15  # Avoid log(0)
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.sum(y_true * np.log(y_pred))

# Example: Cat image
y_true = np.array([1.0, 0.0, 0.0])  # True: Cat
y_pred_good = np.array([0.9, 0.08, 0.02])  # Good prediction
y_pred_bad = np.array([0.3, 0.5, 0.2])  # Bad prediction

loss_good = cross_entropy_loss(y_true, y_pred_good)
loss_bad = cross_entropy_loss(y_true, y_pred_bad)

print('CROSS ENTROPY LOSS')
print('='*60)
print(f'True label: Cat (1, 0, 0)')
print(f'\nGood prediction: {y_pred_good}')
print(f'Loss: {loss_good:.4f}')
print(f'\nBad prediction: {y_pred_bad}')
print(f'Loss: {loss_bad:.4f}')
print('='*60)
print('Lower loss = Better prediction!')
````
