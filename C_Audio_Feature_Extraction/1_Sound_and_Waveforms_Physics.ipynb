## 1. ðŸŒŠ Physics of Sound

Sound is a **mechanical wave** that propagates through a transmission medium (air, water, solids) by particle vibration.

### Key Concepts:

#### ðŸ”¹ **Propagation**
Sound occurs via **particle interaction** - when one particle vibrates, it causes neighboring particles to vibrate, creating a wave that travels through the medium.

#### ðŸ”¹ **Compression & Rarefaction**
- **Compression**: High pressure regions where particles are clustered together
- **Rarefaction**: Low pressure regions where particles are spread apart

These alternating regions of high and low pressure create the sound wave that travels through the medium.

````python
# Visualize sound wave propagation with compression and rarefaction
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8))

# Create a sound wave
x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x)

# Plot 1: Sine wave representation
ax1.plot(x, y, linewidth=3, color='#2E86AB')
ax1.axhline(y=0, color='black', linestyle='--', alpha=0.3)
ax1.fill_between(x, 0, y, where=(y > 0), alpha=0.3, color='red', label='Compression (High Pressure)')
ax1.fill_between(x, 0, y, where=(y < 0), alpha=0.3, color='blue', label='Rarefaction (Low Pressure)')
ax1.set_xlabel('Distance/Time', fontsize=12, fontweight='bold')
ax1.set_ylabel('Pressure', fontsize=12, fontweight='bold')
ax1.set_title('Sound Wave: Compression and Rarefaction', fontsize=14, fontweight='bold')
ax1.legend(fontsize=11)
ax1.grid(True, alpha=0.3)

# Plot 2: Particle displacement visualization
# Simulate particle positions
num_particles = 50
particle_positions = np.linspace(0, 4*np.pi, num_particles)
displacement = 0.3 * np.sin(particle_positions)

for i, pos in enumerate(particle_positions):
    # Color based on compression/rarefaction
    color = 'red' if displacement[i] > 0 else 'blue'
    alpha = 0.3 + 0.7 * abs(displacement[i]) / 0.3
    ax2.scatter(pos, displacement[i], s=100, color=color, alpha=alpha, edgecolors='black', linewidth=0.5)

ax2.axhline(y=0, color='black', linestyle='-', linewidth=2, label='Equilibrium Position')
ax2.set_xlabel('Position', fontsize=12, fontweight='bold')
ax2.set_ylabel('Particle Displacement', fontsize=12, fontweight='bold')
ax2.set_title('Particle Displacement in Sound Wave', fontsize=14, fontweight='bold')
ax2.legend(fontsize=11)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nðŸ“Œ Key Insight:")
print("   â€¢ Red regions (above axis): COMPRESSION - particles pushed together")
print("   â€¢ Blue regions (below axis): RAREFACTION - particles pulled apart")
print("   â€¢ The wave travels through the medium via these alternating pressure zones")
````

## 2. ðŸ“Š The Waveform (Time Domain)

A **waveform** is a visual representation of how a signal varies over time. In audio, we typically plot **amplitude** (y-axis) against **time** (x-axis).

### Mathematical Representation:

A simple sinusoidal wave can be described by:

$$y(t) = A \cdot \sin(2\pi f t)$$

Where:
- $y(t)$ = amplitude at time $t$
- $A$ = maximum amplitude (loudness)
- $f$ = frequency (pitch)
- $t$ = time

````python
# Demonstrate waveform with different parameters
fig, axes = plt.subplots(3, 1, figsize=(14, 10))

# Time array
t = np.linspace(0, 2, 2000)  # 2 seconds

# Waveform 1: Low frequency, low amplitude
f1, A1 = 2, 0.5  # 2 Hz, amplitude 0.5
y1 = A1 * np.sin(2 * np.pi * f1 * t)
axes[0].plot(t, y1, linewidth=2, color='#A23B72', label=f'f={f1} Hz, A={A1}')
axes[0].set_ylabel('Amplitude', fontsize=11, fontweight='bold')
axes[0].set_title('Low Frequency, Low Amplitude (Quiet, Low Pitch)', fontsize=12, fontweight='bold')
axes[0].legend(fontsize=10)
axes[0].grid(True, alpha=0.3)
axes[0].set_ylim(-1.5, 1.5)

# Waveform 2: High frequency, low amplitude
f2, A2 = 8, 0.5  # 8 Hz, amplitude 0.5
y2 = A2 * np.sin(2 * np.pi * f2 * t)
axes[1].plot(t, y2, linewidth=2, color='#F18F01', label=f'f={f2} Hz, A={A2}')
axes[1].set_ylabel('Amplitude', fontsize=11, fontweight='bold')
axes[1].set_title('High Frequency, Low Amplitude (Quiet, High Pitch)', fontsize=12, fontweight='bold')
axes[1].legend(fontsize=10)
axes[1].grid(True, alpha=0.3)
axes[1].set_ylim(-1.5, 1.5)

# Waveform 3: High frequency, high amplitude
f3, A3 = 8, 1.2  # 8 Hz, amplitude 1.2
y3 = A3 * np.sin(2 * np.pi * f3 * t)
axes[2].plot(t, y3, linewidth=2, color='#C73E1D', label=f'f={f3} Hz, A={A3}')
axes[2].set_xlabel('Time (seconds)', fontsize=11, fontweight='bold')
axes[2].set_ylabel('Amplitude', fontsize=11, fontweight='bold')
axes[2].set_title('High Frequency, High Amplitude (Loud, High Pitch)', fontsize=12, fontweight='bold')
axes[2].legend(fontsize=10)
axes[2].grid(True, alpha=0.3)
axes[2].set_ylim(-1.5, 1.5)

plt.tight_layout()
plt.show()

print("\nðŸŽ¯ Formula Breakdown: y(t) = A Â· sin(2Ï€ft)")
print("   â€¢ A (Amplitude): Controls LOUDNESS")
print("   â€¢ f (Frequency): Controls PITCH (how many cycles per second)")
print("   â€¢ t (Time): The independent variable")
print("   â€¢ 2Ï€: Converts frequency to angular frequency (radians/second)")
````

## 3. ðŸŽ›ï¸ Key Parameters of Sound

Understanding these fundamental parameters is crucial for audio signal processing:

### 1ï¸âƒ£ **Amplitude (A)**
- **Definition**: Maximum displacement from equilibrium
- **Perception**: Loudness/Volume
- **Measurement**: Decibels (dB)
- **Range**: 0 (silence) to ~120 dB (pain threshold)

### 2ï¸âƒ£ **Frequency (f)**
- **Definition**: Number of cycles per second
- **Unit**: Hertz (Hz) = cycles/second
- **Perception**: Pitch
- **Formula**: $f = \frac{1}{T}$ (inverse of period)

### 3ï¸âƒ£ **Period (T)**
- **Definition**: Duration of one complete cycle
- **Unit**: Seconds
- **Formula**: $T = \frac{1}{f}$
- **Example**: 440 Hz â†’ T = 1/440 â‰ˆ 0.00227 seconds

### 4ï¸âƒ£ **Phase (Ï†)**
- **Definition**: Offset of the wave relative to time zero
- **Unit**: Degrees (Â°) or Radians
- **Impact**: Crucial for constructive/destructive interference
- **Formula**: $y(t) = A \cdot \sin(2\pi f t + \phi)$

### 5ï¸âƒ£ **Hearing Range**
- **Human**: 20 Hz - 20,000 Hz (20 kHz)
- **Varies by age**: High frequency hearing decreases with age
- **Speech**: Primarily 300 Hz - 3,400 Hz
- **Music**: Typically 20 Hz - 15,000 Hz



````python
# Visualize the relationship between frequency and period
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

t = np.linspace(0, 1, 1000)

# Different frequencies
frequencies = [1, 2, 5, 10]
colors = ['#E63946', '#F77F00', '#06A77D', '#4361EE']

for idx, (f, color) in enumerate(zip(frequencies, colors)):
    ax = axes[idx // 2, idx % 2]
    y = np.sin(2 * np.pi * f * t)
    period = 1 / f
    
    ax.plot(t, y, linewidth=2.5, color=color)
    
    # Mark one complete period
    ax.axvspan(0, period, alpha=0.2, color=color)
    ax.axvline(x=period, color='red', linestyle='--', linewidth=2, label=f'Period T = {period:.2f}s')
    
    ax.set_xlabel('Time (seconds)', fontsize=10, fontweight='bold')
    ax.set_ylabel('Amplitude', fontsize=10, fontweight='bold')
    ax.set_title(f'Frequency = {f} Hz | Period = {period:.3f} s', fontsize=11, fontweight='bold')
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.3)
    ax.set_xlim(0, 1)

plt.tight_layout()
plt.show()

print("\nðŸ“ Frequency-Period Relationship:")
print("   â€¢ Higher frequency â†’ Shorter period (more cycles in same time)")
print("   â€¢ Lower frequency â†’ Longer period (fewer cycles in same time)")
print("   â€¢ Formula: T = 1/f  or  f = 1/T")
````

## 4. ðŸŽ¹ Pitch: Logarithmic Perception

### Why Logarithmic?

Human perception of pitch is **logarithmic**, not linear. This means:

- **Doubling frequency = Going up one octave** (perceived as same interval)
- 220 Hz â†’ 440 Hz = One octave (A3 â†’ A4)
- 440 Hz â†’ 880 Hz = One octave (A4 â†’ A5)
- The frequency **doubles**, but the perceived pitch increase is the **same**

# Deep Dive: Pitch Perception and Logarithmic Scaling

This section explains the fundamental relationship between **Physical Frequency (Physics)** and **Perceived Pitch (Psychophysics)**, and why we cannot simply feed raw linear frequencies into a machine learning model.

---

### 1. The Core Conflict: Linear vs. Logarithmic

*   **Computer/Physics:** Sound is measured in **Hertz (Hz)**. This is a **Linear Scale**.
*   **Human Brain:** Sound is measured in **Pitch**. This is a **Logarithmic Scale**.

#### The "Interval" Experiment
Imagine a tone generator playing two different jumps in frequency:

1.  **Jump A:** Play $100 \text{ Hz}$, then $200 \text{ Hz}$.
2.  **Jump B:** Play $1000 \text{ Hz}$, then $1100 \text{ Hz}$.

| Metric | Jump A (100 $\to$ 200) | Jump B (1000 $\to$ 1100) |
| :--- | :--- | :--- |
| **Physics (Hz)** | Added **100 Hz**. | Added **100 Hz**. |
| **Perception** | Heard an **Octave** (Massive leap). | Heard a **Tiny Step** (Barely noticeable). |

**Conclusion:** Humans do not hear the *difference* ($+$) between frequencies. We hear the **ratio** ($\times$) between frequencies.

---

### 2. The Weber-Fechner Law

This phenomenon is described by the **Weber-Fechner Law**, which states that human sensation is proportional to the logarithm of the physical stimulus intensity.

#### The Relationship
$$ P \propto \log(f) $$

Where:
*   $P$ is Perceived Pitch.
*   $f$ is Frequency in Hz.

To make the pitch sound like it is rising linearly (1, 2, 3 steps), the frequency must rise **exponentially** (100, 200, 400 Hz).

---

### 3. The Biological Reason: The Cochlea

The reason is hardware-based. Inside the ear, the **Cochlea** contains the **Basilar Membrane**.

*   **Base of Membrane:** Stiff/Thick $\rightarrow$ Resonates with **High Frequencies**.
*   **Apex of Membrane:** Floppy/Thin $\rightarrow$ Resonates with **Low Frequencies**.

**The Logarithmic Mapping:**
The physical distance on the membrane between $100 \text{ Hz}$ and $200 \text{ Hz}$ is roughly the **same physical distance** as between $2000 \text{ Hz}$ and $4000 \text{ Hz}$. The brain maps "physical distance on tissue" to "pitch difference."

---

### 4. Musical Implications: The Octave

The foundational unit of music, the **Octave**, represents a frequency ratio of **2:1**.

*   **A1:** $55 \text{ Hz}$
*   **A2:** $110 \text{ Hz}$ (Gap: $+55$)
*   **A3:** $220 \text{ Hz}$ (Gap: $+110$)
*   **A4:** $440 \text{ Hz}$ (Gap: $+220$) $\leftarrow$ *Standard Tuning*
*   **A5:** $880 \text{ Hz}$ (Gap: $+440$)

**The Problem for AI:**
In a linear spectrogram ($0$ to $22000 \text{ Hz}$):
1.  Musical content (Bass, Vocals) exists mostly in $50 \text{ Hz} - 1000 \text{ Hz}$. This is squashed into the **bottom 5%** of the image.
2.  The top 50% of the image ($11 \text{ kHz} - 22 \text{ kHz}$) contains almost no meaningful information ("air" or noise).

---

### 5. The Engineering Solution: The Mel Scale

To fix this for Machine Learning, we warp the frequency axis to match human hearing. We want a scale where the distance between $100-200 \text{ Hz}$ looks the same as $1000-2000 \text{ Hz}$.

#### The Mel Formula
$$ m = 2595 \cdot \log_{10}\left(1 + \frac{f}{700}\right) $$

**How it transforms data:**
*   **< 1000 Hz:** The curve is roughly linear (High resolution for bass/mids).
*   **> 1000 Hz:** The curve becomes logarithmic (Compresses high frequencies).

---


````python
import numpy as np
import matplotlib.pyplot as plt

# Mel scale conversion formulas
def hz_to_mel(hz):
    """Convert frequency in Hz to Mel scale"""
    return 2595 * np.log10(1 + hz / 700)

def mel_to_hz(mel):
    """Convert Mel scale to frequency in Hz"""
    return 700 * (10**(mel / 2595) - 1)

# Create frequency range
hz_range = np.linspace(0, 8000, 1000)  # 0 to 8000 Hz
mel_range = hz_to_mel(hz_range)

# Create Mel range for inverse
mel_values = np.linspace(0, 3000, 1000)  # 0 to 3000 Mel
hz_from_mel = mel_to_hz(mel_values)

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Hz to Mel conversion
axes[0, 0].plot(hz_range, mel_range, linewidth=3, color='blue')
axes[0, 0].set_xlabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Mel Scale', fontsize=12, fontweight='bold')
axes[0, 0].set_title('Hz to Mel Conversion\nMel = 2595 Ã— logâ‚â‚€(1 + Hz/700)', fontsize=14, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].axhline(y=1000, color='red', linestyle='--', alpha=0.5, label='1000 Mel')
axes[0, 0].axvline(x=1000, color='green', linestyle='--', alpha=0.5, label='1000 Hz')
axes[0, 0].legend()

# Plot 2: Mel to Hz conversion
axes[0, 1].plot(mel_values, hz_from_mel, linewidth=3, color='red')
axes[0, 1].set_xlabel('Mel Scale', fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[0, 1].set_title('Mel to Hz Conversion\nHz = 700 Ã— (10^(Mel/2595) - 1)', fontsize=14, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].axhline(y=1000, color='green', linestyle='--', alpha=0.5, label='1000 Hz')
axes[0, 1].axvline(x=1000, color='red', linestyle='--', alpha=0.5, label='1000 Mel')
axes[0, 1].legend()

# Plot 3: Comparison on same scale
axes[1, 0].plot(hz_range, hz_range, linewidth=3, color='gray', linestyle='--', label='Linear (Hz)', alpha=0.5)
axes[1, 0].plot(hz_range, mel_range, linewidth=3, color='blue', label='Mel Scale')
axes[1, 0].set_xlabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Scale Value', fontsize=12, fontweight='bold')
axes[1, 0].set_title('Linear vs Mel Scale Comparison', fontsize=14, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].legend(fontsize=11)

# Plot 4: Mel filter bank spacing
num_filters = 10
mel_points = np.linspace(hz_to_mel(0), hz_to_mel(8000), num_filters + 2)
hz_points = mel_to_hz(mel_points)

axes[1, 1].stem(hz_points, np.ones_like(hz_points), linefmt='blue', markerfmt='bo', basefmt='gray')
axes[1, 1].set_xlabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Filter Center', fontsize=12, fontweight='bold')
axes[1, 1].set_title('Mel Filter Bank Spacing (10 filters)\nEqual spacing in Mel â†’ Non-linear in Hz', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3, axis='x')
axes[1, 1].set_ylim(0, 1.5)

# Add annotations showing spacing
for i in range(len(hz_points)):
    axes[1, 1].text(hz_points[i], 1.1, f'{int(hz_points[i])}', 
                    ha='center', fontsize=9, rotation=45)

plt.tight_layout()
plt.savefig('mel_scale_visualization.png', dpi=300, bbox_inches='tight')
plt.show()

# Print some example conversions
print("\n=== Example Conversions ===")
test_frequencies = [100, 500, 1000, 2000, 4000, 8000]
print("\nHz â†’ Mel:")
for hz in test_frequencies:
    mel = hz_to_mel(hz)
    print(f"  {hz:5d} Hz = {mel:7.2f} Mel")

print("\nMel â†’ Hz:")
test_mels = [100, 500, 1000, 1500, 2000, 2500]
for mel in test_mels:
    hz = mel_to_hz(mel)
    print(f"  {mel:4d} Mel = {hz:7.2f} Hz")
````

````python
import numpy as np
import matplotlib.pyplot as plt

def hz_to_mel(hz):
    """Convert frequency in Hz to mel scale"""
    return 2595 * np.log10(1 + hz / 700)

# Generate frequency range from 20 Hz to 20,000 Hz
frequencies = np.logspace(np.log10(20), np.log10(20000), 1000)
log_frequencies = np.log10(frequencies)
mel_values = hz_to_mel(frequencies)

# Create linear fit for comparison (above 1000 Hz)
mask = frequencies >= 1000
linear_fit = np.polyfit(log_frequencies[mask], mel_values[mask], 1)
linear_approx = np.polyval(linear_fit, log_frequencies)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(log_frequencies, mel_values, linewidth=2, color='blue', label='Actual Mel Scale')
plt.plot(log_frequencies, linear_approx, '--', linewidth=2, color='red', alpha=0.7, label='Linear Approximation (fit >1kHz)')
plt.xlabel('Log10(Frequency) [Hz]', fontsize=12)
plt.ylabel('Mel Scale', fontsize=12)
plt.title('Log Frequency vs Mel Scale (Linear Relationship) - is not Linear for initial/lower frequencies', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# Demonstrate logarithmic perception of pitch
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Generate octaves starting from A0
A0 = 27.5  # Hz
octaves = np.arange(0, 9)
frequencies = A0 * (2 ** octaves)
note_names = [f'A{i}' for i in octaves]

# Plot 1: Linear scale
ax1.bar(note_names, frequencies, color='#E63946', alpha=0.7, edgecolor='black', linewidth=2)
ax1.set_xlabel('Musical Note', fontsize=12, fontweight='bold')
ax1.set_ylabel('Frequency (Hz)', fontsize=12, fontweight='bold')
ax1.set_title('Frequency on LINEAR Scale\n(Exponential growth)', fontsize=13, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')

# Add frequency labels on bars
for i, (note, freq) in enumerate(zip(note_names, frequencies)):
    ax1.text(i, freq + 100, f'{freq:.1f} Hz', ha='center', fontsize=9, fontweight='bold')

# Plot 2: Logarithmic scale
ax2.bar(note_names, frequencies, color='#06A77D', alpha=0.7, edgecolor='black', linewidth=2)
ax2.set_xlabel('Musical Note', fontsize=12, fontweight='bold')
ax2.set_ylabel('Frequency (Hz) - Log Scale', fontsize=12, fontweight='bold')
ax2.set_title('Frequency on LOGARITHMIC Scale\n(Equal spacing = Equal perception)', fontsize=13, fontweight='bold')
ax2.set_yscale('log')
ax2.grid(True, alpha=0.3, axis='y')

# Add frequency labels on bars
for i, (note, freq) in enumerate(zip(note_names, frequencies)):
    ax2.text(i, freq * 1.15, f'{freq:.1f} Hz', ha='center', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

print("\nðŸŽµ Key Insight: Logarithmic Perception")
print("   â€¢ LEFT: Linear scale shows exponential growth")
print("   â€¢ RIGHT: Log scale shows equal spacing (how we perceive it)")
print("   â€¢ Each octave doubles the frequency, but sounds like the same interval")
print(f"\n   Frequency progression: {' â†’ '.join([f'{f:.1f} Hz' for f in frequencies[:6]])}")
print("   Each step doubles the previous frequency!")
````

## 5. ðŸ“ˆ Pitch-Frequency Relationship & Mapping

### The Mathematical Formula

The relationship between MIDI pitch number and frequency is given by:

$$F(p) = 2^{\frac{p-69}{12}} \cdot 440$$

Where:
- $F(p)$ = Frequency in Hz
- $p$ = MIDI pitch number (0-127)
- $69$ = MIDI number for A4 (440 Hz reference)
- $12$ = Number of semitones per octave
- $440$ = Reference frequency (A4 in Hz)

### Breaking Down the Formula:

1. **$(p - 69)$**: Distance in semitones from A4
2. **$\frac{p-69}{12}$**: Distance in octaves from A4
3. **$2^{\frac{p-69}{12}}$**: Frequency multiplier (doubling per octave)
4. **$\cdot 440$**: Scale to Hz using A4 reference

### Examples:
- **A4** (p=69): $F(69) = 2^{\frac{69-69}{12}} \cdot 440 = 2^0 \cdot 440 = 440$ Hz âœ“
- **A5** (p=81): $F(81) = 2^{\frac{81-69}{12}} \cdot 440 = 2^1 \cdot 440 = 880$ Hz âœ“
- **A3** (p=57): $F(57) = 2^{\frac{57-69}{12}} \cdot 440 = 2^{-1} \cdot 440 = 220$ Hz âœ“

````python
import numpy as np
import matplotlib.pyplot as plt

# Mel scale conversion formulas
def hz_to_mel(hz):
    """Convert frequency in Hz to Mel scale"""
    return 2595 * np.log10(1 + hz / 700)

def mel_to_hz(mel):
    """Convert Mel scale to frequency in Hz"""
    return 700 * (10**(mel / 2595) - 1)

# Create frequency range
hz_range = np.linspace(0, 8000, 1000)  # 0 to 8000 Hz
mel_range = hz_to_mel(hz_range)

# Create Mel range for inverse
mel_values = np.linspace(0, 3000, 1000)  # 0 to 3000 Mel
hz_from_mel = mel_to_hz(mel_values)

# Create figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot 1: Hz to Mel conversion
axes[0, 0].plot(hz_range, mel_range, linewidth=3, color='blue')
axes[0, 0].set_xlabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Mel Scale', fontsize=12, fontweight='bold')
axes[0, 0].set_title('Hz to Mel Conversion\nMel = 2595 Ã— logâ‚â‚€(1 + Hz/700)', fontsize=14, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3)
axes[0, 0].axhline(y=1000, color='red', linestyle='--', alpha=0.5, label='1000 Mel')
axes[0, 0].axvline(x=1000, color='green', linestyle='--', alpha=0.5, label='1000 Hz')
axes[0, 0].legend()

# Plot 2: Mel to Hz conversion
axes[0, 1].plot(mel_values, hz_from_mel, linewidth=3, color='red')
axes[0, 1].set_xlabel('Mel Scale', fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[0, 1].set_title('Mel to Hz Conversion\nHz = 700 Ã— (10^(Mel/2595) - 1)', fontsize=14, fontweight='bold')
axes[0, 1].grid(True, alpha=0.3)
axes[0, 1].axhline(y=1000, color='green', linestyle='--', alpha=0.5, label='1000 Hz')
axes[0, 1].axvline(x=1000, color='red', linestyle='--', alpha=0.5, label='1000 Mel')
axes[0, 1].legend()

# Plot 3: Comparison on same scale
axes[1, 0].plot(hz_range, hz_range, linewidth=3, color='gray', linestyle='--', label='Linear (Hz)', alpha=0.5)
axes[1, 0].plot(hz_range, mel_range, linewidth=3, color='blue', label='Mel Scale')
axes[1, 0].set_xlabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[1, 0].set_ylabel('Scale Value', fontsize=12, fontweight='bold')
axes[1, 0].set_title('Linear vs Mel Scale Comparison', fontsize=14, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)
axes[1, 0].legend(fontsize=11)

# Plot 4: Mel filter bank spacing
num_filters = 10
mel_points = np.linspace(hz_to_mel(0), hz_to_mel(8000), num_filters + 2)
hz_points = mel_to_hz(mel_points)

axes[1, 1].stem(hz_points, np.ones_like(hz_points), linefmt='blue', markerfmt='bo', basefmt='gray')
axes[1, 1].set_xlabel('Frequency (Hz)', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Filter Center', fontsize=12, fontweight='bold')
axes[1, 1].set_title('Mel Filter Bank Spacing (10 filters)\nEqual spacing in Mel â†’ Non-linear in Hz', fontsize=14, fontweight='bold')
axes[1, 1].grid(True, alpha=0.3, axis='x')
axes[1, 1].set_ylim(0, 1.5)

# Add annotations showing spacing
for i in range(len(hz_points)):
    axes[1, 1].text(hz_points[i], 1.1, f'{int(hz_points[i])}', 
                    ha='center', fontsize=9, rotation=45)

plt.tight_layout()
plt.savefig('mel_scale_visualization.png', dpi=300, bbox_inches='tight')
plt.show()

# Print some example conversions
print("\n=== Example Conversions ===")
test_frequencies = [100, 500, 1000, 2000, 4000, 8000]
print("\nHz â†’ Mel:")
for hz in test_frequencies:
    mel = hz_to_mel(hz)
    print(f"  {hz:5d} Hz = {mel:7.2f} Mel")

print("\nMel â†’ Hz:")
test_mels = [100, 500, 1000, 1500, 2000, 2500]
for mel in test_mels:
    hz = mel_to_hz(mel)
    print(f"  {mel:4d} Mel = {hz:7.2f} Hz")
````

## 6. ðŸŽ¼ Cents: Fine-Tuning Pitch

### What are Cents?

**Cents** are a logarithmic unit of measure used for musical intervals. They provide a way to express very small differences in pitch.

### Key Facts:

1. **One Octave = 1200 cents**
   - Divides the octave into 1200 equal parts
   - Logarithmic scale (like decibels for loudness)

2. **One Semitone = 100 cents**
   - 12 semitones per octave
   - 12 Ã— 100 = 1200 cents per octave

3. **Noticeable Pitch Difference: 10-25 cents**
   - Trained musicians can detect ~5-10 cents
   - Average listeners notice ~25 cents
   - Context-dependent (easier to hear in sustained notes)

### Mathematical Formula:

The number of cents between two frequencies:

$$\text{cents} = 1200 \times \log_2\left(\frac{f_2}{f_1}\right)$$

### Why Cents?

- **Tuning**: Fine-tune instruments (e.g., "tune up 5 cents")
- **Analysis**: Measure intonation accuracy
- **Comparison**: Compare different tuning systems
- **Perception**: Quantify just-noticeable differences

````python
# Visualize cents and their perceptual significance
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Function to calculate cents
def cents_difference(f1, f2):
    """Calculate the difference in cents between two frequencies"""
    return 1200 * np.log2(f2 / f1)

# Plot 1: Cents in one octave
cents_in_octave = np.linspace(0, 1200, 13)
semitone_names = ['C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#', 'A', 'A#', 'B', 'C']

axes[0, 0].bar(range(13), cents_in_octave, color='#E63946', alpha=0.7, edgecolor='black', linewidth=2)
axes[0, 0].set_xticks(range(13))
axes[0, 0].set_xticklabels(semitone_names, fontsize=10, fontweight='bold')
axes[0, 0].set_ylabel('Cents from Starting Note', fontsize=11, fontweight='bold')
axes[0, 0].set_title('Cents Distribution Across One Octave (Each semitone = 100 cents)', fontsize=12, fontweight='bold')
axes[0, 0].grid(True, alpha=0.3, axis='y')

for i, cents in enumerate(cents_in_octave):
    axes[0, 0].text(i, cents + 30, f'{int(cents)}Â¢', ha='center', fontsize=9, fontweight='bold')

# Plot 2: Perceptual thresholds
thresholds = {
    'Imperceptible': (0, 5),
    'Trained Ear': (5, 10),
    'Noticeable': (10, 25),
    'Clearly Off': (25, 50),
    'Very Off': (50, 100),
}

colors_thresh = ['#2ECC71', '#F39C12', '#E67E22', '#E74C3C', '#8B0000']
y_pos = 0

for (label, (low, high)), color in zip(thresholds.items(), colors_thresh):
    width = high - low
    axes[0, 1].barh(y_pos, width, left=low, height=0.6, color=color, alpha=0.7, edgecolor='black', linewidth=2)
    axes[0, 1].text(low + width/2, y_pos, f'{label}\n{low}-{high}Â¢', 
                   ha='center', va='center', fontsize=10, fontweight='bold', color='white',
                   bbox=dict(boxstyle='round,pad=0.4', facecolor='black', alpha=0.7))
    y_pos += 1

axes[0, 1].set_xlabel('Cents Deviation', fontsize=11, fontweight='bold')
axes[0, 1].set_title('Perceptual Thresholds for Pitch Deviation', fontsize=12, fontweight='bold')
axes[0, 1].set_yticks([])
axes[0, 1].set_xlim(0, 100)
axes[0, 1].grid(True, alpha=0.3, axis='x')

# Plot 3: Frequency ratio to cents conversion
ratios = np.linspace(1, 2, 100)
cents_from_ratio = 1200 * np.log2(ratios)

axes[1, 0].plot(ratios, cents_from_ratio, linewidth=3, color='#06A77D')
axes[1, 0].set_xlabel('Frequency Ratio (fâ‚‚/fâ‚)', fontsize=11, fontweight='bold')
axes[1, 0].set_ylabel('Cents', fontsize=11, fontweight='bold')
axes[1, 0].set_title('Frequency Ratio to Cents Conversion (1:1 to 2:1 = 0 to 1200 cents)', fontsize=12, fontweight='bold')
axes[1, 0].grid(True, alpha=0.3)

# Mark important ratios
important_ratios = [(1.0, 0, 'Unison'), (1.0595, 100, 'Semitone'), (1.5, 702, 'Perfect 5th'), (2.0, 1200, 'Octave')]

for ratio, cents, label in important_ratios:
    axes[1, 0].plot(ratio, cents, 'ro', markersize=10, markeredgecolor='black', markeredgewidth=1.5)
    axes[1, 0].text(ratio, cents + 50, f'{label}\n{ratio:.3f}:{cents:.0f}Â¢', 
                   ha='center', fontsize=8, fontweight='bold',
                   bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))

# Plot 4: Practical example - A4 detuning
A4 = 440
cents_detune = np.linspace(-50, 50, 101)
frequencies_detune = A4 * (2 ** (cents_detune / 1200))

axes[1, 1].plot(cents_detune, frequencies_detune, linewidth=3, color='#4361EE')
axes[1, 1].axhline(y=440, color='red', linestyle='--', linewidth=2, label='Perfect A4 (440 Hz)')
axes[1, 1].axvline(x=0, color='red', linestyle='--', linewidth=2, alpha=0.5)
axes[1, 1].axvspan(-10, 10, alpha=0.2, color='green', label='Barely Noticeable (Â±10Â¢)')
axes[1, 1].axvspan(-25, -10, alpha=0.2, color='orange')
axes[1, 1].axvspan(10, 25, alpha=0.2, color='orange', label='Noticeable (Â±10-25Â¢)')

axes[1, 1].set_xlabel('Cents Deviation from A4', fontsize=11, fontweight='bold')
axes[1, 1].set_ylabel('Frequency (Hz)', fontsize=11, fontweight='bold')
axes[1, 1].set_title('A4 Detuning: Frequency vs Cents Deviation', fontsize=12, fontweight='bold')
axes[1, 1].legend(fontsize=9)
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nðŸŽ¯ Cents Summary:")
print("   â€¢ 1 Octave = 1200 cents")
print("   â€¢ 1 Semitone = 100 cents")
print("   â€¢ Noticeable difference: 10-25 cents (context-dependent)")
print("   â€¢ Formula: cents = 1200 Ã— logâ‚‚(fâ‚‚/fâ‚)")
print("\n   Example: 440 Hz to 880 Hz (one octave)")
print(f"   cents = 1200 Ã— logâ‚‚(880/440) = 1200 Ã— logâ‚‚(2) = 1200 Ã— 1 = 1200Â¢")
````
